## 광운대 모델

```python
result_li = glob('/content/drive/MyDrive/rPPG/광운대/results/*.csv')
```

```python
MAE_ARIMA = []
MAE_ARIMA_MAF = []
MAE_ARIMA_SF = []
RMSE_ARIMA = []
RMSE_ARIMA_MAF = []
RMSE_ARIMA_SF = []
PCC_ARIMA = []
PCC_ARIMA_MAF = []
PCC_ARIMA_SF = []

for i in result_li:
    temp_df = pd.read_csv(i,index_col=0)

    temp_ARIMA_mae = mean_absolute_error(temp_df['bpm'],temp_df['bpm_ARIMA'])
    temp_ARIMA_MAF_mae = mean_absolute_error(temp_df['bpm'],temp_df['bpm_ARIMA_MAF'])
    temp_ARIMA_SF_mae = mean_absolute_error(temp_df['bpm'],temp_df['bpm_ARIMA_SF'])

    temp_ARIMA_rmse = mean_squared_error(temp_df['bpm'],temp_df['bpm_ARIMA'])**0.5
    temp_ARIMA_MAF_rmse = mean_squared_error(temp_df['bpm'],temp_df['bpm_ARIMA_MAF'])**0.5
    temp_ARIMA_SF_rmse = mean_squared_error(temp_df['bpm'],temp_df['bpm_ARIMA_SF'])**0.5

    temp_ARIMA_pcc = pearsonr(temp_df['bpm'],temp_df['bpm_ARIMA'])[0]
    temp_ARIMA_MAF_pcc = pearsonr(temp_df['bpm'],temp_df['bpm_ARIMA_MAF'])[0]
    temp_ARIMA_SF_pcc = pearsonr(temp_df['bpm'],temp_df['bpm_ARIMA_SF'])[0]

    MAE_ARIMA.append(temp_ARIMA_mae)
    MAE_ARIMA_MAF.append(temp_ARIMA_MAF_mae)
    MAE_ARIMA_SF.append(temp_ARIMA_SF_mae)
    RMSE_ARIMA.append(temp_ARIMA_rmse)
    RMSE_ARIMA_MAF.append(temp_ARIMA_MAF_rmse)
    RMSE_ARIMA_SF.append(temp_ARIMA_SF_rmse)
    PCC_ARIMA.append(temp_ARIMA_pcc)
    PCC_ARIMA_MAF.append(temp_ARIMA_MAF_pcc)
    PCC_ARIMA_SF.append(temp_ARIMA_SF_pcc)

MAE_ARIMA_mean = np.mean(MAE_ARIMA)
MAE_ARIMA_MAF_mean = np.mean(MAE_ARIMA_MAF)
MAE_ARIMA_SF_mean = np.mean(MAE_ARIMA_SF)
RMSE_ARIMA_mean = np.mean(RMSE_ARIMA)
RMSE_ARIMA_MAF_mean = np.mean(RMSE_ARIMA_MAF)
RMSE_ARIMA_SF_mean = np.mean(RMSE_ARIMA_SF)
PCC_ARIMA_mean = np.mean(PCC_ARIMA)
PCC_ARIMA_MAF_mean = np.mean(PCC_ARIMA_MAF)
PCC_ARIMA_SF_mean = np.mean(PCC_ARIMA_SF)
```

```python
data = {}

## 기존 pyVHR 라이브러리
data['CHROM_clustering_mean'] = CHROM_clustering_mean
data['POS_clustering_mean'] = POS_clustering_mean
data['LGI_clustering_mean'] = LGI_clustering_mean
data['ICA_clustering_mean'] = ICA_clustering_mean
data['PCA_clustering_mean'] = PCA_clustering_mean

data['CHROM_holistic_mean'] = CHROM_holistic_mean
data['POS_holistic_mean'] = POS_holistic_mean
data['LGI_holistic_mean'] = LGI_holistic_mean
data['ICA_holistic_mean'] = ICA_holistic_mean
data['PCA_holistic_mean'] = PCA_holistic_mean

data['CHROM_median_mean'] = CHROM_median_mean
data['POS_median_mean'] = POS_median_mean
data['LGI_median_mean'] = LGI_median_mean
data['ICA_median_mean'] = ICA_median_mean
data['PCA_median_mean'] = PCA_median_mean

data['MTTS_CAN_deep_mean'] = MTTS_CAN_deep_mean
data['HR_CNN_deep_mean'] = HR_CNN_deep_mean

## 광운대 모델
data['광운대_ARIMA'] = [MAE_ARIMA_mean, RMSE_ARIMA_mean, PCC_ARIMA_mean]
data['광운대_ARIMA_MAF'] = [MAE_ARIMA_MAF_mean, RMSE_ARIMA_MAF_mean, PCC_ARIMA_MAF_mean]
data['광운대_ARIMA_SF'] = [MAE_ARIMA_SF_mean, RMSE_ARIMA_SF_mean, PCC_ARIMA_SF_mean]
```

## 결과

```python
res_df = pd.DataFrame(data).set_index(pd.Index([
    'MAE','RMSE','PCC',
    # ,'CCC', 'SNR', 'MAX'
    ]))
res_df.to_csv('/content/drive/MyDrive/rPPG/광운대/results/result_integ.csv')
```
